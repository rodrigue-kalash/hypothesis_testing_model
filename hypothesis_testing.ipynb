# Transform non Normal distribution into Normal distribuation

1st Method Log Transformation

```python

x_log = pd.DataFrame()
for col in x :
  log_l = []
  if (col =="color"or  col == "pH"):
    continue
  else:
    for d in x[col]:
      log_l.append(np.log1p(d))
    x_log["log_"+col] = log_l
```

```python

x_log.head()
```

```python
def hist_qq ( x, col):
  plt.figure(figsize = (15,6))
  plt.subplot(1,2,1)
  x[col].hist()
  plt.title(col)
  plt.subplot(1,2,2)
  stats.probplot(x[col],dist="norm",plot=plt)
  plt.show()
```

```python
for col in x_log:
  hist_qq(x_log,col)
```

2sd Method Reciprocal Method

```python

x_rec = pd.DataFrame()
for col in x :
  rec_l = []
  if (col =="color"or  col == "pH"):
    continue
  else:
    for d in x[col]:
      rec_l.append(1/(d+1))
    x_rec["recipro_"+col] = rec_l
```

```python
x_rec.head()
```

```python
for col in x_rec:
  hist_qq(x_rec,col)
```

3rd Method Square Root Transformation

```python

x_sqroot = pd.DataFrame()
for col in x :
  sqr_l = []
  if (col =="color"or  col == "pH"):
    continue
  else:
    for d in x[col]:
      sqr_l.append(d**(0.5))
    x_sqroot["sqrt_"+col] = sqr_l
```

```python
x_sqroot.head()
```

```python
for col in x_sqroot:
  hist_qq(x_sqroot,col)
```

4th Method: Exponential Transformation

```python
x_exp = pd.DataFrame()
for col in x :
  exp_l = []
  if (col =="color"or  col == "pH"):
    continue
  else:
    for d in x[col]:
      exp_l.append(np.exp(d))
    x_exp["exp_"+col] = exp_l
```

```python
x_exp.head()
```

```python
for col in x_exp:
  hist_qq(x_exp,col)
```
